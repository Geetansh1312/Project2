```python
from datetime import date
from jugaad_data.nse import stock_df
df = stock_df(symbol="SBIN", from_date=date(2020,1,1),
            to_date=date(2020,1,30), series="EQ")
```

```python
df
```

```python
df['daily_returns']=None
```

```python
df.drop(columns='daily_returns',axis=1,inplace=True)
```

```python
df
```

```python
df['daily_returns']=(df['CLOSE']-df['PREV. CLOSE'])/df['PREV. CLOSE']
```

```python
df
```

```python
sum=0
c=0
for x in df['CLOSE']:
    sum=sum+x
    c=c+1
sma=sum/c
```

```python
sma
```

```python
lis1=[]
sum1=0
c1=0
for x in range(21,-1,-1):
    
    for x1 in range(21,x-1,-1):
        sum1=df.loc[x1,'CLOSE']+sum1
        c1=c1+1
    lis1.append(sum1/c1)
    sum1=0
    c1=0
lis2=lis1[::-1]
        
        
```

```python
df.drop('MA',axis=1,inplace=True)
```

```python
df['MA']=lis2
```

```python
df
```

```python
k=2/23
```

```python
cp=df.loc[21,'CLOSE']
```

```python
cp
```

```python
ema=cp
k=2/23
lis2=[cp]
for x in range(20,-1,-1):
    ema=(df.loc[x,'CLOSE']*k)+(df.loc[x+1,'CLOSE']*(1-k))
    lis2.append(ema)


    
    
```

```python
lis3=lis2[::-1]
```

```python
df['EMA']=lis3
```

```python
df
```

```python
avg_daily_return=0
for x in range(0,22):
    avg_daily_return+=df.loc[x,'daily_returns']
avg_daily_return/=22

 
```

```python
avg_daily_return
```

```python
import math
volatility=[]
v=0
c=0
for x in range(21,-1,-1):
    c+=1
    v=math.sqrt(math.pow((df.loc[x,'daily_returns']-avg_daily_return),2)/c)
    volatility.append(v)
df['volatility']=volatility[::-1]
    
    
```
