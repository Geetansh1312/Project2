```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
np.random.seed(123)
data_similar = {
    'Age': np.random.randint(18, 70, 350),
    'Height': np.random.randint(140, 200, 350),
    'Weight': np.random.randint(50, 100, 350),
}

# Convert to DataFrame
df_similar = pd.DataFrame(data_similar)

# Introduce some missing values (similar to the previous example)
df_similar.loc[df_similar.sample(frac=0.1).index, 'Age'] = np.nan
df_similar.loc[df_similar.sample(frac=0.05).index, 'Height'] = np.nan

# Introduce outliers (similar to the previous example)
outliers_similar = pd.DataFrame({
    'Age': [16, 75],
    'Height': [220, 130],
    'Weight': [40, 120],
})
df_similar = pd.concat([df_similar, outliers_similar], ignore_index=True)
```

```python
df_similar
```

```python
df=df_similar
```

```python
df
```

```python
median_imputer=SimpleImputer(missing_values=np.nan,strategy='median')
median_imputer=median_imputer.fit(df)
imputed_df=median_imputer.transform(df)
df1=pd.DataFrame(data=imputed_df,columns=df.columns)
```
